```python
import os
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.preprocessing.image import load_img, img_to_array

# Define your image directory
image_dir = 'data/training_data/'

# Read image filenames and labels
image_filenames = []
labels = []

for label in os.listdir(image_dir):
    label_dir = os.path.join(image_dir, label)
    print(label)
    for filename in os.listdir(label_dir):
        image_filenames.append(os.path.join(label_dir, filename))
        labels.append(label)

# Load images and convert to arrays
images = [img_to_array(load_img(img, target_size=(224, 224))) for img in image_filenames]

# Convert labels to numerical values using LabelEncoder
le = LabelEncoder()
encoded_labels = le.fit_transform(labels)
```

```python

# Create a DataFrame with image paths and labels
df = pd.DataFrame({'image_path': image_filenames, 'label': encoded_labels})

# Display the first few rows of the DataFrame
print(df.head())
```

```python
# y = le.fit_transform(all_label)
classes = list(le.classes_)
from tensorflow.python.keras.utils import np_utils
y=np_utils.to_categorical(encoded_labels,num_classes=len(labels))
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(np.array(images),np.array(y),stratify=y,test_size = 0.2,random_state=777,shuffle=True)
```

```python
def create_dataset(path):
    image_paths=[]
    image_labels=[]
    folders = os.listdir(path)
    labels = '\n'.join(sorted(folders))
    with open('labels.txt','w')as f:
        f.write(labels)
    for dir in os.listdir(path):
        for file in os.listdir(f'{path}/{dir}'):
            image_path= f'{path}/{dir}/{file}'
            image_label = 0 if dir == 'cats' else 1
            image_paths.append(image_path)
            image_labels.append(image_label)
    image_paths=np.array(image_paths)
    image_labels=np.array(image_labels)
    return image_paths , image_labels
```

```python
x_train.shape,y_train.shape,x_test.shape,y_test.shape
```

```python
model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Flatten())
model.add(tf.keras.layers.Dense(128,activation=tf.nn.relu))
model.add(tf.keras.layers.Dense(128,activation=tf.nn.relu))
model.add(tf.keras.layers.Dense(10,activation=tf.nn.softmax))
```

```python
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(x=x_train,y=y_train,batch_size=32,epochs=10) 
```

```python
val_loss , val_acc = model.evaluate(x_test,y_test)
print(val_loss , val_acc )
```

```python
plt.plot(model.history.history['accuracy'])
plt.plot(model.history.history['val_accuracy'])
plt.title('diget - train')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['accuracy','val_accuracy'],loc='upper left')
plt.show()

plt.plot(model.history.history['loss'])
plt.plot(model.history.history['val_loss'])
plt.title('diget - train')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['loss','val_loss'],loc='upper left')
plt.show()
```
